# Logestic Regression

```python
n= 100
x= np.linspace (0, 10, n)

y= np.array([0]*int(n/2) + [1]*int(n/2))

plt.plot (x, y, 'o')
plt.plot (x[[0, 90]], y[[0,90]])
```

```python
theta1= 10
theta0= 5
z= theta1*(x-theta0)
ypred= 1/(1+np.exp(-z)) 

plt.plot (x, y, 'o')

plt.plot (x, ypred)
```

Using Scikit Learn

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression ()
model.fit (x.reshape (-1,1), y)
```

```python
ypred1= model.predict (x.reshape (-1,1))
```

```python
plt.plot (x, y, 'o')
plt.plot (x, ypred1)
```

```python
x_test= np.linspace (-5, 15, 500)
ypred2= model.predict (x_test.reshape (-1,1))
```

```python
plt.plot (x, y, 'o')
plt.plot (x_test, ypred2)
```
